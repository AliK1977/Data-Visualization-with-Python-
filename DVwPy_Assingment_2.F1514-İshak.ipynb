**1b: Swarm Plot**

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
#with seaborn  

fig,ax=plt.subplots(figsize = (10,6))

sns.swarmplot(x="gender", y="age", hue="manner_of_death",data=kill)
plt.show()
```

**2: Plot 12 most dangerous cities**

```python
city_freq = kill.city.value_counts().head(12)
city_freq
```

```python
a=city_freq.index
a
```

```python
b=city_freq.values
b
```

```python
# with seaborn barplot

fig, ax =plt.subplots(figsize=(8,5))
ax = sns.barplot(x=a,y=b)

ax =plt.xticks(rotation=45)
ax =plt.title("Most Dangerous Cities") 
plt.show()
```

```python
# with seaborn countplot

plt.figure(figsize=(8,5))
sns.countplot(x="city",order=a, data=kill)

plt.xticks(rotation=45)
plt.title('Most Dangerous Cities')
plt.show()
```

**3: Plot 20 most dangerous states**

```python
k=state_freq.index
k
```

```python
# with seaborn barplot

plt.figure(figsize=(8,5))

sns.barplot(x=k, y=m)

plt.xticks(rotation=45)
plt.title('Most Dangerous States')
plt.show()
```
